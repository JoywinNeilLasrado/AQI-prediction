```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
csv_file_path = r"C:\Users\Subhiksha Rai K\Desktop\Jupyter\DV\AQI and Lat Long of Countries.csv" # Update with the correct path
aqi_data = pd.read_csv(csv_file_path)

# Set theme for visualizations
sns.set_theme(style="whitegrid")
```

```python
# Facet Grid: AQI Value Distribution by Category
g = sns.FacetGrid(aqi_data, col="AQI Category", col_wrap=4, height=4, sharex=False)
g.map(sns.histplot, "AQI Value", bins=20, color="green")
g.fig.suptitle("AQI Value Distribution by AQI Category", y=1.05, fontsize=16)
g.set_axis_labels("AQI Value", "Frequency")
plt.tight_layout()
plt.show()
```
