```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
csv_file_path = r"AQI and Lat Long of Countries.csv" # Update with the correct path
aqi_data = pd.read_csv(csv_file_path)

# Set theme for visualizations
sns.set_theme(style="whitegrid")
```

```python
# Plot: Distribution of AQI Categories
plt.figure(figsize=(10, 6))
sns.countplot(y="AQI Category", data=aqi_data, order=aqi_data["AQI Category"].value_counts().index)
plt.title("Distribution of AQI Categories", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("AQI Category", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Select relevant AQI value columns
aqi_values = aqi_data[["AQI Value", "CO AQI Value", "Ozone AQI Value", "NO2 AQI Value", "PM2.5 AQI Value"]]

# Plot: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(aqi_values.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of AQI Values", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Plot: Geospatial Scatterplot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(aqi_data['lng'], aqi_data['lat'], c=aqi_data['AQI Value'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='AQI Value')
plt.title("Geospatial Distribution of AQI Values", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Box Plot: AQI Value by AQI Category
plt.figure(figsize=(12, 6))
sns.boxplot(x="AQI Category", y="AQI Value", data=aqi_data, palette="Set2")
plt.title("Box Plot of AQI Values by AQI Category", fontsize=16)
plt.xlabel("AQI Category", fontsize=12)
plt.ylabel("AQI Value", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram of AQI Values
plt.figure(figsize=(10, 6))
sns.histplot(aqi_data["AQI Value"], bins=30, kde=True, color="blue")
plt.title("Histogram of AQI Values", fontsize=16)
plt.xlabel("AQI Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Pair Plot of AQI Values
sns.pairplot(aqi_data[["CO AQI Value", "Ozone AQI Value", "NO2 AQI Value", "PM2.5 AQI Value"]], diag_kind="kde", palette="husl")
plt.suptitle("Pair Plot of AQI Pollutants", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Swarm Plot: PM2.5 AQI Value vs AQI Category
plt.figure(figsize=(12, 6))
sns.swarmplot(x="AQI Category", y="PM2.5 AQI Value", data=aqi_data, palette="muted")
plt.title("PM2.5 AQI Value by AQI Category", fontsize=16)
plt.xlabel("AQI Category", fontsize=12)
plt.ylabel("PM2.5 AQI Value", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Facet Grid: AQI Value Distribution by Category
g = sns.FacetGrid(aqi_data, col="AQI Category", col_wrap=4, height=4, sharex=False)
g.map(sns.histplot, "AQI Value", bins=20, color="green")
g.fig.suptitle("AQI Value Distribution by AQI Category", y=1.05, fontsize=16)
g.set_axis_labels("AQI Value", "Frequency")
plt.tight_layout()
plt.show()
```
